1.2 Complete a função abaixo para extrair uma coluna, do arquivo csv em uma lista.


```python
def extrai_coluna_csv(nome_arquivo: str, indice_coluna: int) -> str:

  coluna = list()

  try:
      
      with open(file=nome_arquivo, mode='r', encoding='utf8') as fp:
           linha = fp.readline()
           linha = fp.readline()
           while linha:
                 linha_separada = linha.split(sep=',')
                 valor = linha_separada[indice_coluna]
                 coluna.append(valor)
                 linha = fp.readline()

  except Exception as exc:
         print(exc)

  return coluna
```

## 1\. Funções para arquivo csv

Complete a função abaixo para extrair uma coluna do arquivo csv em uma lista. Os elementos devem ter o tipo de dado correto.

```python
def extrai_coluna_csv(nome_arquivo: str, indice_coluna: int, tipo_dado: str) -> str:

  coluna = list()

  try:
      
      with open(file=nome_arquivo, mode='r', encoding='utf8') as fp:
           linha = fp.readline()
           linha = fp.readline()
           while linha:
                 linha_separada = linha.split(sep=',')
                 valor = linha_separada[indice_coluna]
                 if type(valor) == type(tipo_dado):
                    coluna.append(valor)
                    linha = fp.readline()
                 else:
                    valor = type(valor)
                    coluna.append(valor)
                    linha = fp.readline()

  except Exception as exc:
         print(exc)

  return coluna
```

Você pode testar a função com os códigos abaixo.

```python
# extrair a coluna valor_venda
valor_venda = extrai_coluna_csv(nome_arquivo='./carros.csv', indice_coluna=1, tipo_dado='str')
print(valor_venda) # deve retornar ['vhigh', 'med', 'low', ...]
```

```python
# extrair a coluna pessoas
pessoas = extrai_coluna_csv(nome_arquivo='./carros.csv', indice_coluna=4, tipo_dado='int')
print(pessoas) # deve retornar [2, 2, 2, ...]
```

```python
def extrai_linha_txt(nome_arquivo: str, numero_linha: int):

  palavras = list()

  try:
      
      with open(file=nome_arquivo, mode='r', encoding='utf8') as fp:
           linha = fp.readlines()
           valor = linha[numero_linha-1]
           palavras = valor.split(sep=' ')
 
  except Exception as exc:
         print(exc)

  return palavras
```
